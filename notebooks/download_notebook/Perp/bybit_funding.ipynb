# Bybit — Funding Rate: Extraction and 1h Alignment

This notebook builds the funding-rate dataset used in the empirical analysis. It retrieves funding observations from Bybit’s REST API, standardizes timestamps in UTC, and produces an hourly series designed to be merged on `ts` with the OHLCV (klines) and open-interest datasets.

## Context
Funding is typically observed at discrete publication times (product-dependent). Downstream analysis in this project operates on an **hourly grid**, so this notebook converts the raw funding observations into a 1-hour indexed series that can be joined consistently with other hourly panels.

## Source and time conventions
- **Provider:** Bybit
- **Interface:** REST API (funding-related endpoint)
- **Time standard:** all timestamps are handled in **UTC**
- **Window convention:** extraction covers **[START, END_EXCL)** (end bound is *exclusive*)

## Inputs (set in the configuration cell)
- `SYMBOL` (e.g., `ETHUSDT`)
- `CATEGORY` (market type used in the study design, e.g., `linear`)
- `START` and `END_EXCL` (UTC; `END_EXCL` is exclusive)

## What the notebook does (in order)
1) **Setup**: reads parameters, defines the time window, and prepares output paths under `data/`.
2) **Retrieval**: queries the API over the requested window (paged/segmented requests, with basic retry logic).
3) **Normalization**: converts raw payloads into a tidy table (typed columns, UTC datetimes, sorted and de-duplicated).
4) **Hourly alignment**: maps the funding observations onto an hourly index used by the rest of the pipeline.
5) **QA and export**: writes raw, processed, QA, and a manifest with run parameters and file hashes.

## Outputs
Files are written under `data/`:

- `data/raw/...jsonl` — raw API payloads (audit trail)
- `data/processed/...parquet` — hourly funding series (analysis-ready)
- `data/qa/..._qa.json` — QA summary (gaps, duplicates, bounds, missingness)
- `data/manifests/..._manifest.json` — parameters and SHA256 hashes of the outputs

## Hourly alignment (method note)
Funding is observed at discrete timestamps. To work on a common hourly panel, the notebook constructs an hourly time index and assigns a funding value to each hour using the project’s alignment rule (e.g., carrying forward the most recent available funding observation until the next update). This choice should be kept consistent across the empirical pipeline so merges on `ts` are well-defined.

## Processed dataset (schema)
One row per hour (UTC):
- `ts` (datetime UTC; hourly index)
- `funding_rate` (float)
- `symbol`, `category`
- `source` (e.g., `"bybit"`)
- `ingested_at_utc` (datetime UTC)

Additional columns may be present if the API returns them (kept as metadata rather than inputs to the merge key).

## Reproducibility note
The pipeline is reproducible given the same code and parameters. Exact bit-for-bit reproducibility depends on the provider returning identical historical data for the same query window. The manifest hashes record the exact artifacts produced for each run.

## Quick notes
- Use the QA JSON to check time bounds, duplicates, and missing hours before debugging the API retrieval.
- The extraction window is **[START, END_EXCL)**; adjacent runs can be concatenated without overlap.
- When merging with OHLCV/open interest, join on `ts` (UTC) and keep the same hourly alignment rule throughout the project.

```python
import json, time, hashlib, random
from pathlib import Path
from typing import Dict, Any, List, Tuple, Optional

import pandas as pd
import numpy as np
import requests
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry
```

```python
# ===== CONFIG =====
OUT_DIR = Path("raw_data_lake") / "bybit_funding"
BASE_URL = "https://api.bybit.com"

CATEGORY = "linear"
SYMBOL = "ETHUSDT"

START = pd.Timestamp("2021-03-15T00:00:00Z")
END_EXCL = pd.Timestamp("2025-12-01T00:00:00Z")

# Bybit funding history constraints (IMPORTANT)
LIMIT = 200              # max 200 on funding endpoint

# HTTP
MIN_REQUEST_INTERVAL_S = 0.6
TIMEOUT = (10, 120)
RETRIES_TOTAL = 8
BACKOFF_FACTOR = 1.0

# pagination
CHUNK_DAYS = 365
MAX_PAGES_PER_CHUNK = 20000

# funding grid policy
FUNDING_FILL: Optional[str] = None   # None (raw) or "ffill"

OUT_DIR.mkdir(parents=True, exist_ok=True)
print("OUT_DIR =", OUT_DIR)
print("Range:", START, "->", END_EXCL, "(end exclusive)")
print("Funding fill:", FUNDING_FILL)
```

```python
def ms(ts: pd.Timestamp) -> int:
    return int(ts.value // 10**6)

def sha256_file(path: Path) -> str:
    h = hashlib.sha256()
    with path.open("rb") as f:
        for chunk in iter(lambda: f.read(1024 * 1024), b""):
            h.update(chunk)
    return h.hexdigest()

def append_jsonl(path: Path, obj: Dict[str, Any]) -> None:
    with path.open("a", encoding="utf-8") as f:
        f.write(json.dumps(obj, ensure_ascii=False) + "\n")

def utc_hour_grid(start: pd.Timestamp, end_excl: pd.Timestamp) -> pd.DatetimeIndex:
    return pd.date_range(start=start, end=end_excl, freq="h", inclusive="left", tz="UTC")

def now_utc_iso() -> str:
    return pd.Timestamp.utcnow().isoformat()
```

```python
def make_session() -> requests.Session:
    s = requests.Session()
    retry = Retry(
        total=RETRIES_TOTAL,
        backoff_factor=BACKOFF_FACTOR,
        status_forcelist=(429, 500, 502, 503, 504),
        allowed_methods=("GET",),
        raise_on_status=False,
    )
    s.mount("https://", HTTPAdapter(max_retries=retry))
    return s

SESSION = make_session()
_last_req_t = 0.0

def _pace():
    global _last_req_t
    now = time.monotonic()
    wait = MIN_REQUEST_INTERVAL_S - (now - _last_req_t)
    if wait > 0:
        time.sleep(wait + random.uniform(0.0, 0.15))
    _last_req_t = time.monotonic()

def bybit_get(path: str, params: Dict[str, Any]) -> Dict[str, Any]:
    url = f"{BASE_URL}{path}"

    for attempt in range(1, RETRIES_TOTAL + 1):
        _pace()
        t0 = time.time()
        r = SESSION.get(url, params=params, timeout=TIMEOUT)
        elapsed = time.time() - t0

        try:
            data = r.json()
        except Exception:
            data = {"retCode": None, "retMsg": "Non-JSON response", "_http_status": r.status_code}

        if data.get("retCode") == 10006:
            sleep_s = min(60.0, (2 ** (attempt - 1)) * 0.8 + random.uniform(0.0, 0.8))
            print(f"[rate_limit] 10006 attempt={attempt}/{RETRIES_TOTAL} sleep={sleep_s:.2f}s")
            time.sleep(sleep_s)
            continue

        return {
            "response": data,
            "http_status": r.status_code,
            "elapsed_s": elapsed,
            "url": url,
        }

    raise RuntimeError("Rate limit persists after retries.")
```

```python
def fetch_funding_raw(start: pd.Timestamp, end_excl: pd.Timestamp) -> Tuple[pd.DataFrame, Dict[str, Any]]:
    pages_path = OUT_DIR / "pages.jsonl"
    if pages_path.exists():
        pages_path.unlink()

    all_items: List[dict] = []
    requests_made = stalls = empty_pages = 0

    chunk_start = start
    while chunk_start < end_excl:
        chunk_end = min(chunk_start + pd.Timedelta(days=CHUNK_DAYS), end_excl)
        print(f"[funding] chunk {chunk_start:%Y-%m-%d} -> {chunk_end:%Y-%m-%d}")

        cursor_end = chunk_end
        prev_sig = None
        safety = 0

        while True:
            safety += 1
            if safety > MAX_PAGES_PER_CHUNK:
                raise RuntimeError("Too many pages in chunk; reduce CHUNK_DAYS.")

            params = {
                "category": CATEGORY,
                "symbol": SYMBOL,
                "startTime": ms(chunk_start),
                "endTime": ms(cursor_end),
                "limit": LIMIT,
            }

            page = bybit_get("/v5/market/funding/history", params)
            res = page["response"]
            requests_made += 1

            append_jsonl(pages_path, {
                "t": time.time(),
                "path": "/v5/market/funding/history",
                "params": params,
                "http_status": page["http_status"],
                "elapsed_s": page["elapsed_s"],
                "retCode": res.get("retCode"),
                "retMsg": res.get("retMsg"),
            })

            if res.get("retCode") != 0:
                raise RuntimeError(f"Bybit retCode={res.get('retCode')} retMsg={res.get('retMsg')}")

            lst = (res.get("result") or {}).get("list") or []
            if not lst:
                empty_pages += 1
                break

            # timestamp field (usually fundingRateTimestamp)
            ts_field = "fundingRateTimestamp" if "fundingRateTimestamp" in lst[0] else None
            if ts_field is None:
                for k in lst[0].keys():
                    if "Timestamp" in k:
                        ts_field = k
                        break
            if ts_field is None:
                raise ValueError(f"Cannot find funding timestamp field. Keys={list(lst[0].keys())}")

            ts = [int(x[ts_field]) for x in lst]
            page_min_ms, page_max_ms = min(ts), max(ts)
            sig = (len(lst), page_min_ms, page_max_ms)
            if prev_sig is not None and sig == prev_sig:
                stalls += 1
                print("[funding][warn] repeated page signature -> break chunk")
                break
            prev_sig = sig

            all_items.extend(lst)

            page_min = pd.to_datetime(page_min_ms, unit="ms", utc=True)
            if page_min <= chunk_start:
                break

            new_cursor_end = page_min - pd.Timedelta(milliseconds=1)
            if new_cursor_end >= cursor_end:
                stalls += 1
                print("[funding][warn] cursor not moving -> break chunk")
                break
            cursor_end = new_cursor_end

        chunk_start = chunk_end

    df = pd.DataFrame(all_items)

    summary = {
        "requests_made": requests_made,
        "rows_raw": int(len(all_items)),
        "stalls": stalls,
        "empty_pages": empty_pages,
    }
    return df, summary
```

```python
def normalize_funding_1h_ffill(
    df_raw: pd.DataFrame,
    start: pd.Timestamp,
    end_excl: pd.Timestamp,
):
    """
    Normalize Bybit funding history to an hourly grid using forward-fill.

    Interpretation:
    - funding_rate is observed at settlement times (typically every 8h)
    - the last known funding_rate is assumed constant until the next settlement

    This version is merge-ready with 1h klines.
    """

    # Hourly grid
    idx = utc_hour_grid(start, end_excl)

    if df_raw.empty:
        out = pd.DataFrame({"date": idx, "funding_rate": np.nan})
        qa = {
            "expected_hours": int(len(idx)),
            "rows_raw_after_dedup": 0,
            "missing_funding": int(out["funding_rate"].isna().sum()),
            "note": "empty raw funding input",
        }
        return out, qa

    df = df_raw.copy()

    # Robust column detection
    ts_col = next(
        (c for c in ["fundingRateTimestamp","fundingRateTime","fundingTime","timestamp","time"] if c in df.columns),
        None
    )
    rate_col = next(
        (c for c in ["fundingRate","funding_rate","rate"] if c in df.columns),
        None
    )
    if ts_col is None or rate_col is None:
        raise ValueError(f"Funding columns missing. Have={list(df.columns)}")

    # Parse + align to hour
    df[ts_col] = pd.to_numeric(df[ts_col], errors="coerce")
    df["date"] = pd.to_datetime(df[ts_col], unit="ms", utc=True).dt.floor("h")
    df["funding_rate"] = pd.to_numeric(df[rate_col], errors="coerce")

    # Deduplicate (keep last if Bybit repeats)
    df = df.sort_values("date").drop_duplicates("date", keep="last")

    # Clip to study window
    df = df[(df["date"] >= start) & (df["date"] < end_excl)].copy()

    # Merge on hourly grid
    out = (
        pd.DataFrame({"date": idx})
        .merge(df[["date","funding_rate"]], on="date", how="left")
        .sort_values("date")
    )

    # Forward-fill funding between settlements
    out["funding_rate"] = out["funding_rate"].ffill()

    qa = {
        "expected_hours": int(len(idx)),
        "rows_raw_after_dedup": int(len(df)),
        "missing_funding": int(out["funding_rate"].isna().sum()),
        "funding_min": float(out["funding_rate"].min(skipna=True)) if out["funding_rate"].notna().any() else None,
        "funding_max": float(out["funding_rate"].max(skipna=True)) if out["funding_rate"].notna().any() else None,
        "method": "hourly grid + forward-fill between funding settlements",
    }

    return out, qa
```

```python
df_raw, http_summary = fetch_funding_raw(START, END_EXCL)
df_norm, qa = normalize_funding_1h_ffill(df_raw, START, END_EXCL)

raw_csv = OUT_DIR / "funding_raw.csv"
raw_parq = OUT_DIR / "funding_raw.parquet"
norm_parq = OUT_DIR / "funding_1h.parquet"
qa_path = OUT_DIR / "qa_report.json"

df_raw.to_csv(raw_csv, index=False)
df_raw.to_parquet(raw_parq, index=False, engine="pyarrow")

df_norm.to_parquet(norm_parq, index=False, engine="pyarrow")
qa_path.write_text(json.dumps(qa, indent=2), encoding="utf-8")

manifest = {
    "dataset": "bybit_funding",
    "collected_at_utc": now_utc_iso(),
    "config": {
        "base_url": BASE_URL,
        "category": CATEGORY,
        "symbol": SYMBOL,
        "start": str(START),
        "end_excl": str(END_EXCL),
        "limit": LIMIT,
        "chunk_days": CHUNK_DAYS,
        "min_request_interval_s": MIN_REQUEST_INTERVAL_S,
        "timeout": TIMEOUT,
        "retries_total": RETRIES_TOTAL,
        "backoff_factor": BACKOFF_FACTOR,
        "funding_fill": FUNDING_FILL,
    },
    "files": {
        "raw_csv": str(raw_csv),
        "raw_parquet": str(raw_parq),
        "norm_parquet": str(norm_parq),
        "pages_jsonl": str(OUT_DIR / "pages.jsonl"),
        "qa_report": str(qa_path),
    },
    "hashes": {
        "raw_csv_sha256": sha256_file(raw_csv),
        "raw_parquet_sha256": sha256_file(raw_parq),
        "norm_parquet_sha256": sha256_file(norm_parq),
        "qa_report_sha256": sha256_file(qa_path),
    },
    "http_summary": http_summary,
    "qa": qa,
}

(OUT_DIR / "manifest.json").write_text(json.dumps(manifest, indent=2), encoding="utf-8")

print("✅ Saved:", norm_parq)
print("HTTP summary:", http_summary)
print("QA:", qa)

df_norm.head(12)
```
